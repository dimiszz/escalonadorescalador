```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('out.csv')
#display(df)

#GRAFICOS INDIVIDUAIS MEDIA TROCAS E INSTRUCAO
plt.plot(df['QUANTUMS'], df['MEDIA_TROCAS'])
plt.xlabel('Quantidade de Quantum')
plt.ylabel('Média de trocas')
plt.title('Relação entre Média de Trocas e Quantum')
plt.show()

plt.plot(df['QUANTUMS'], df['MEDIA_INSTRUCOES'])
plt.xlabel('Quantidade de Quantum')
plt.ylabel('Média de Instruções')
plt.title('Relação entre Média de Instruções e Quantum')
plt.show()

df['DER _MEDIA_INSTRUCOES'] = df['MEDIA_INSTRUCOES'].diff()
df['DER _MEDIA_TROCAS'] = df['MEDIA_TROCAS'].diff()
#display(df)

#GRAFICOS INDIVIDUAIS DERIVADAS DAS MEDIAS TROCAS E INSTRUCOES
plt.plot(df['QUANTUMS'], df['DER _MEDIA_INSTRUCOES'])
plt.xlabel('Quantidade de Quantum')
plt.ylabel('Média de Instruções')
plt.title('Taxa de variação entre Média de Instruções e Quantum')
plt.show()

plt.plot(df['QUANTUMS'], df['DER _MEDIA_TROCAS'])
plt.xlabel('Quantidade de Quantum')
plt.ylabel('Média de trocas')
plt.title('Taxa de variação entre Média de Trocas e Quantum')
plt.show()

#GRAFICOS CONJUNTOS DAS MEDIAS TROCAS E INSTRUCOES
plt.plot(df['QUANTUMS'], df['MEDIA_INSTRUCOES'], color='blue', label='Der_media')
plt.plot(df['QUANTUMS'], df['MEDIA_TROCAS'], color='red', label='Der_trocas')

# Adicionar rótulos e título
plt.title('Relação - Média de Instruções e Trocas x Quantums')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Adicionar uma legenda para identificar as linhas
plt.legend()

# Exibir o gráfico
plt.show()


#GRAFICOS CONJUNTOS DERIVADA DAS MEDIAS TROCAS E INSTRUCOES
plt.plot(df['QUANTUMS'], df['DER _MEDIA_INSTRUCOES'], color='blue', label='Der_media')
plt.plot(df['QUANTUMS'], df['DER _MEDIA_TROCAS'], color='red', label='Der_trocas')

# Adicionar rótulos e título
plt.title('Taxa de Variação - Média de Instruções e Trocas')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Adicionar uma legenda para identificar as linhas
plt.legend()

# Exibir o gráfico
plt.show()




df_pares = df[df.index % 2 != 0]
df_impares = df[df.index % 2 == 0]

#display(df_impares)
#display(df_pares)

#GRAFICOS CONJUNTOS DERIVADA DAS MEDIAS TROCAS E INSTRUCOES
plt.plot(df_impares['QUANTUMS'], df_impares['DER _MEDIA_INSTRUCOES'], color='blue', label='Der_impares')
plt.plot(df_pares['QUANTUMS'], df_pares['DER _MEDIA_INSTRUCOES'], color='red', label='Der_pares')

# Adicionar rótulos e título
plt.title('Taxa de Variação - Média de Instruções com Quantumspares e impares')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Adicionar uma legenda para identificar as linhas
plt.legend()

# Exibir o gráfico
plt.show()
```
